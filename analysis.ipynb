```python
x=np.array([1,2,6,8,3])
y=np.array([7,8,9,5,3])
plt.figure(figsize=(7,5),dpi=150,edgecolor='red')
plt.plot(x,y,'o',color='orange',label='line 1')
plt.plot(x+2,y-2,'.g',label='line 2')
plt.plot(x+4,y+2,'>',label='line 3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='lower right',bbox_to_anchor=(0.8, 0., 0.5, 0.5))
plt.show()
```

```python
x=np.array([1,2,6,8,3])
y=np.array([7,8,9,5,3])
plt.bar(x,y)
plt.plot(y+4)
```
