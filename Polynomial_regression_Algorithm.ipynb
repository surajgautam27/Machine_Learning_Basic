```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
```

```python
data_baseball=pd.read_csv('/content/drive/MyDrive/data/baseball.csv')
```

```python
data_baseball
```

```python
X=data_baseball.drop(columns=['player','salary'])
y=data_baseball['salary']
```

```python
data_baseball.info()
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Lasso_reg=Lasso(alpha=0.3)
```

```python
Lasso_reg.fit(X_train,y_train)
```

```python
Lasso_reg.score(X_train,y_train)
```

```python
Lasso_reg.score(X_test,y_test)
```

```python
Lasso_reg.coef_
```
