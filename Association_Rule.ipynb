```python
transaction=[]
for i in range(dataset_original.shape[0]):
  transaction.append([str(dataset_original.values[i,j]) for j in range (dataset_original.shape[1])])
transaction= np.array(transaction)
transaction
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transaction).transform(transaction)
dataset_f=pd.DataFrame(te_ary, columns=te.columns_)
dataset_f
```

```python
first30 = df_table["Items"].head(120).values
dataset = dataset_f.loc[:,first30]
dataset.shape
```

```python
dataset.shape
```

```python
df_table["Items"]
```

```python
from mlxtend.frequent_patterns import apriori
res=apriori(dataset, min_support=0.05,use_colnames=True , verbose=2)
res
```

```python
from mlxtend.frequent_patterns import association_rules
rule1= association_rules(res, metric='lift' , min_threshold=1)
rule1
```

```python
rule2= association_rules(res, metric='confidence' , min_threshold=0.2)
rule2
```

```python
rule1.sort_values('lift', ascending=False)
```
