```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/Titanic.csv')
data
```

```python
data.info()
```

```python
data=pd.get_dummies(data,columns=['sex','embarked','class','who'])
```

```python
data
```

```python
data['family_size']=1+data['sibsp']+data['parch']
```

```python
data=data.drop(columns=['sibsp','parch'])
```

```python
data
```

```python
data.info()
```

```python
X= data.drop(columns=['survived'])
y=data['survived']
```

```python
X.isna().sum()
```

```python
X['age'].fillna(X['age'].mean(), inplace=True)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.svm import SVC
clf = SVC(verbose=True)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.score(X_test,y_test)
```

```python
clf.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = {'C':[0.01,0.1,1,10,100,1000],
        'kernel' : ["linear","poly","rbf","sigmoid"],
        'degree' : [1,3,5],
        'gamma' : ['scale',0.01,1]
        }
```

```python
grid_2 = {'C':[0.1,1,10],
        'kernel' : ["linear","sigmoid"]
        }
```

```python
svm=SVC(verbose=True)
svm_cv=GridSearchCV(svm,grid_2,cv=5)
svm_cv.fit(X_train,y_train)
print("Best Parameters:",svm_cv.best_params_)
print("Train Score:",svm_cv.best_score_)
```
