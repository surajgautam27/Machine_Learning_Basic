```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris.keys()
```

```python
print(iris.DESCR)
```

```python
X=iris.data[:,[2,3]]
```

```python
X
```

```python
y=iris.target
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0, stratify=y)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
np.unique(y)
```

```python
np.bincount(y),np.bincount(y_train),np.bincount(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_decision = DecisionTreeClassifier(criterion='entropy')
tree_decision.fit(x_train,y_train)
```

```python
tree_decision.score(x_train,y_train)
```

```python
tree_decision.score(x_test,y_test)
```

```python
probab=tree_decision.predict([[5.1,3.5]])
```

```python
iris.target_names[probab]
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train,tree_decision)
plt.xlabel('petal length')
plt.ylabel('petal width')
```

```python
plot_decision_regions(x_test,y_test,tree_decision)
plt.xlabel('petal length')
plt.ylabel('petal width')
```

```python
from sklearn import tree
tree.plot_tree(tree_decision)
```

```python
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
dot_data=export_graphviz(tree_decision,filled=True,class_names=['setosa', 'versicolor', 'virginica'],feature_names=['petal length (cm)', 'petal width (cm)'])
graph=graph_from_dot_data(dot_data)
graph.write_png('iris.png')
```
